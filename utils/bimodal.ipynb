```python
import cv2
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "FreeSerif"
plt.rcParams["font.size"] = 15
```

```python
images = ['../data/egs/ihc_gray.png', '../data/egs/00018_gray.png']
```

```python
fig = cv2.imread(images[0])
plt.figure(figsize=(5,8))
plt.hist(fig.ravel(), 255)
# plt.show()
plt.tight_layout()
plt.savefig('ihc_hist.png')
```

```python
fig = cv2.imread(images[1])
plt.figure(figsize=(5,8))
plt.hist(fig.ravel(), 255)
# plt.show()
plt.tight_layout()
plt.savefig('00018_hist.png')
```
